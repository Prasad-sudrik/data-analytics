```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
gas = pd.read_csv("gas_prices.csv")
```

```python
gas
```

```python
plt.figure(figsize=(10,5), dpi = 700)
plt.plot(gas.Year,gas.Australia,'y.-',label = "Austraila")
plt.plot(gas.Year,gas.Canada,'r.-',label = "Canada")
plt.plot(gas.Year,gas["South Korea"],'g.-',label = "South Korea")
plt.plot(gas.Year,gas["USA"],'b.-',label = "USA")
plt.xticks(gas.Year[::3].to_list() + [2011])
plt.legend()
plt.ylabel('Price per barrel (in USD)',fontdict={'fontsize':'16'})
plt.xlabel("Year" ,fontdict={'fontsize':'16'})
plt.title('Gas Prices per Year' ,fontdict={'fontname':'Comic Sans MS' ,'fontsize':'20'})
plt.savefig('first.png',dpi = 300)
plt.show()
```

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head(5)
```

```python
bins = [40,50,60,70,80,90,100]
plt.figure(figsize=(12,8))
plt.hist(fifa.Overall,bins,color = "#abcdef")
plt.title("Distribution of player SkillS n FIFA 2018")
plt.xticks(bins)
plt.xlabel("Numbers of players")
plt.ylabel("Skill level")
plt.savefig("second.png",dpi= 300)
plt.show()
```
